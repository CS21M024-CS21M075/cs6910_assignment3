### Different Functions

```python
#Softmax Function
def softmax(x):
    total_sum = sum([np.exp(i) for i in x])
    return [np.exp(i) / total_sum for i in x]
```

```python
# Get Colour Function
def color_get(value, mode):
    if(mode == 'l'):
        colors = ['#85c2e1', '#89c4e2', '#95cae5', '#99cce6', '#a1d0e8', '#b2d9ec', '#baddee', '#c2e1f0', '#eff7fb', '#f9e8e8', '#f9e8e8', '#f9d4d4', '#f9bdbd', '#f8a8a8', '#f68f8f', '#f47676', '#f45f5f', '#f34343', '#f33b3b', '#f42e2e']
        value = int((value * 100) / 5)
        return colors[value]
    else:
        colors = ['#ffffff', '#ecf7fb', '#daeff7', '#c7e7f3', '#b5dfef', '#a2d7eb', '#90cfe7', '#7dc7e3', '#6abfdf', '#58b7db', '#46afd7']
        value = int((value * 100) / 10)
        return colors[value]
```

```python
#cstr function
def cstr(s, color = 'black'):
    return "<text style=color:#000;padding-top:1.5px;padding-bottom:1.5px;padding-left:2.5px;padding-right:2.5px;background-color:{}>{} </text>".format(color, s)
```

```python
# Function to Visualize Characters
def visualize_character(decoder_character, text_colours):
    if (decoder_character == "<e>"):
      display(HTML(''.join([cstr(ti, color = ci) for ti, ci in text_colours]) + " <b> &emsp; &lt; e &gt; </b>  &emsp; &nbsp; "))
    else:
      display(HTML(''.join([cstr(ti, color = ci) for ti, ci in text_colours]) + " <b> &emsp; {}</b>  &emsp; &emsp; ".format(decoder_character)))
```

```python
#Function to Visualize Sequence
def visualize_l(decoder_sequence, prob):
    colours = []
    for c, p in zip(decoder_sequence, prob): 
        text = (c, color_get(p, 'l'))
        colours.append(text)
    display(HTML(''.join([cstr(ti, color = ci) for ti, ci in colours])))
```

### Main Connectivity Function

```python
#Taking Value of N = 10
N = 10
with open("/kaggle/input/connectivity/connectivity_visualization.txt", "r", encoding='utf-8') as file:
    #Reading Each Lines i the connectivity_visualization.txt file    
    lines = file.readlines()
    
    
    idx = 0
    visualized_words = 0

    #visualizing connections of each sequences
    while idx < len(lines) and  visualized_words<N:
        line = lines[idx]
            
        #checking next conditions
        if line[:4] == "Next":
            idx += 1
            visualized_words += 1
            continue
        else:
            idx += 1
            actual_word, decoder_character_length = line.split('\t') 

            for j in range(int(decoder_character_length)):
                line = lines[idx]
                line = line.split('\t')
                text_colours = []
                decoder_character = line[0]
                
                #probability array for all the probabilities
                probabilities = []
                for probability_index in range(len(actual_word)) :
                    probabilities.append(float(line[probability_index+1]))

                line = softmax(probabilities)
                    
                for probability_index in range(len(actual_word)) :
                    actual_character = actual_word[probability_index]
                    text= (actual_character, color_get(float(line[probability_index]), 'c') )
                    text_colours.append(text)
        
                visualize_character(decoder_character, text_colours)
                idx += 1
        print()
```
